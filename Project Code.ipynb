```python
# creating mapping from non null 
county_to_area = hospital[hospital['Hospital Service Area'].notna()].groupby('Hospital County')['Hospital Service Area'].agg(lambda x: x.value_counts().index[0])
```

```python
# replacing '120+' to 120 to ensure the column to be in numeric only

hospital['Length of Stay'] = hospital['Length of Stay'].replace('120+', 120)
```

```python
# convert to numeric
hospital['Length of Stay'] = pd.to_numeric(hospital['Length of Stay'], errors='coerce')
```

```python
# Define bin edges and labels
bins = [0, 1, 3, 7, 14, 30, np.inf]
labels = ['Very Short Stay', 'Short Stay', 'Medium Stay', 'Long Stay', 'Very Long Stay', 'Extended Stay']
```

```python

# Create new categorized column
hospital['Length of Stay Category'] = pd.cut(hospital['Length of Stay'], bins=bins, labels=labels, right=True)
```
